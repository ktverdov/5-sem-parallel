```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#n = 10^7, m = 50000

P = (1, 2, 4, 8, 16)

mergesort_serial = 2.1478
qsort = 2.0586
mergesort_openmp = [2.2704, 1.21, 1.109, 0.9628, 0.848]

mergesort_pthreads = np.array([np.mean([2.175, 2.139, 2.136]), np.mean([1.181, 1.163, 1.159]), 
                               np.mean([1.106, 1.118, 1.122]), np.mean([0.880, 0.876, 0.891]),
                               np.mean([0.868, 0.896, 0.892])])
print(mergesort_pthreads)
```

```python
plt.figure(figsize=(15, 7))
plt.hlines(qsort, P[0], P[-1], label='qsort')

plt.plot(P, mergesort_openmp, label='OpenMP')
plt.scatter(P, mergesort_openmp, marker='o', color='red')

plt.plot(P, mergesort_pthreads, label='POSIX Threads')
plt.plot(P, mergesort_pthreads, marker='o', color='red')

plt.title(r'T(P) with $N = 10^7$, $m = 5*10^4$', size=14)
plt.legend()
plt.grid(ls=':')
plt.show()


S = mergesort_serial / mergesort_pthreads

plt.figure(figsize=(15, 7))
plt.plot(P, S)
plt.scatter(P, S, marker='o', color='red')
plt.title(r'S(P) with $N = 10^7$, $m = 5*10^4$', size=14)
plt.grid(ls=':')
plt.show()


E = S / P

plt.figure(figsize=(15, 7))
plt.plot(P, E)
plt.scatter(P, E, marker='o', color='red')
plt.title(r'E(P) with $N = 10^7$, $m = 5*10^4$', size=14)
plt.grid(ls=':')
plt.show()
```
